## 1. Задание
Напишите код на Python, реализующий расчет длины вектора, заданного его координатами. (в программе)

```python
def vector_lenght(x, y, z):
    return math.sqrt(x * x + y * y + z * z)
```

```python
print('Введите координаты вектора:')
x = float(input('X = '))
y = float(input('Y = '))
z = float(input('Z = '))
print(f"Длина вектора {vector_lenght(x, y, z)}")
```

## 3. Задание (в программе)
Напишите код на Python, реализующий построение графиков:
1. окружности,
2. эллипса,
3. гиперболы.


### 1. Окружность

```python
def to_xy_for_circle(r,phi):
     return r*np.cos(phi), r*np.sin(phi)
fig = plt.figure()
axis = fig.add_subplot(111 ,aspect='equal')  
phis = np.arange(0,2*np.pi,0.01)
r = 1.
axis.plot( *to_xy_for_circle(r, phis), color = 'r',linestyle = '-' )
plt.show()
```

### 2. Эллипс

```python
def to_xy_for_ellipse(a, b, phi):
     return a*np.cos(phi), b*np.sin(phi)
fig = plt.figure()
axis = fig.add_subplot(111 ,aspect='equal')  
phis = np.arange(0,2*np.pi,0.01)
a = 3.
b = 2.
axis.plot( *to_xy_for_ellipse(a, b, phis), color = 'r',linestyle = '-' )
plt.show()
```

### 3. Гипербола

```python
x1 = []
x2 = []
y1 = []
y2 = []
k = 5
for i in range(1, 1000):
    x_ = i/1000
    x1.append(x_)
    x2.append(-x_)
    y1.append(k/x_)
    y2.append(k/(-x_))    
plt.plot(x1,y1)
plt.plot(x2,y2)
plt.ylim(-100,100)
plt.xlabel("x")
plt.ylabel("y")
```
