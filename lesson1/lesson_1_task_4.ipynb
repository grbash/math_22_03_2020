```python
%matplotlib inline
import random as rand
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-np.pi, np.pi, 201)
k = rand.randint(2, 20)
print(f"Красным цветом отмечен график y = cos(kx), k равно = {k}")
print("Зеленым цветом отмечен график y = cos(x)")
plt.plot(x, np.cos(x), color='green')
plt.plot(x, np.cos(k*x), color='red')
plt.xlabel('x')
plt.ylabel('y')
plt.axis('tight')
plt.show()
```
